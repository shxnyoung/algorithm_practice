```python
import csv
import folium

f = open('park.csv')
d = csv.reader(f)
next(d) #데이터의 첫 줄 건너뛰기

mymap = folium.Map(location = [37.571412,127.024467],zoom_start =17)
for k in d:
    if '용산' in k[4]:
        folium.Marker([float(k[5]), float(k[6])],popup=k[1]).add_to(mymap)
mymap
```
